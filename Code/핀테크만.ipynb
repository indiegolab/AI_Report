```python
ldamodel,topic_prob = DF.LDA_TF_IDF(15)
news['topic_prob'] = topic_prob
```

```python
def Topic_List():
    # 기사별 토픽 붙이기 
    topic_prob_list = []
    for i in topic_prob:
        tmp = i
        tmp.sort(key=lambda x: x[1], reverse = True)
        topic_prob_list.append(tmp[0][0])  # 토픽 0번~9번

    news['topic'] = topic_prob_list
    return news
```

```python
def Topic_List_Count():  
    import pandas as pd
    new2 = pd.date_range("2019-01-01", "2019-08-31")
    df = pd.DataFrame()
    df['date']= new2

    type(df['date'][0])
    df['date'] = df['date'].apply(lambda x: str(x)[:10])

    df = df.set_index('date')

    new2 = df
    for i in range(10):
        news_topic = news[news['topic']==i]
        news_topic = news_topic.reset_index()
        del news_topic['index']
        news_topic = news_topic.drop_duplicates(['content'])

        news_topic['date'] = news_topic['date'].str[:10]
        news_topic.set_index('date')
        df_count = pd.DataFrame(news_topic['date'].value_counts())
        df_count = df_count.sort_index()
        df_count.index.names = ['date']
        df_count.columns = ['Topic_Count{}'.format(i)]
        print(type(df_count))
        new2 = new2.merge(df_count,how = 'left', on='date')
    
    return new2
        
```
